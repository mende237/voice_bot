```python
class generator:
    def __init__(self):
        self.template = {}
        
        
    def contain(self , tokens , string):
        for token in tokens:
            if token.lower() in string.lower():
                return token
        return ""
        
    def fit(self , data_frame , tokens):
        self.template["PV"] = []
        self.template["TD"] = []
        self.template["cours"] = []
        self.template["requete"] = []
        for i in data_frame.index:
            formulation = data_frame["formulation"][i]
            token = self.contain(tokens , formulation)
            if token != "":
                formulation = formulation.lower().replace(token.lower() , '{code}')
                    
            if "pv".lower() in data_frame["demande"][i].lower() or "pvs".lower() in data_frame["demande"][i].lower(): 
                if token != "":
                    self.template["PV"].append(formulation)
            elif "TD".lower() in data_frame["demande"][i].lower() or "TDs".lower() in data_frame["demande"][i].lower():
                if token != "":
                    self.template["TD"].append(formulation)
            elif "cours".lower() in data_frame["demande"][i].lower():
                if token != "":
                    self.template["cours"].append(formulation)
            elif "requête".lower() in data_frame["demande"][i].lower() or "requêtes".lower() in data_frame["demande"][i].lower():
                if token != "":
                    self.template["requete"].append(formulation)
            
        #print(self.template)
        
    def generate(self , tokens):
        demande = ["pv {code}" , "programmation cours {code}" , "programmation TD {code}" , "requete {code}"]
        keys = ["PV" , "TD" , "cours" , "requete"]
        data = pd.DataFrame(columns = ["demande" , "formulation"])
        cmpt = 0
        for key in keys:    
            formulation_model = self.template[key]
            intent = ""
        
            for token in tokens:
                if key.lower() == "pv".lower():
                    intent = demande[0].format(code = token)
                elif key.lower() == "td".lower():
                    intent = demande[1].format(code = token)
                elif key.lower() == "cours".lower():
                    intent = demande[2].format(code = token)
                elif key.lower() == "requete".lower():
                    intent = demande[3].format(code = token)

                for model in formulation_model:
                    formulation = model.format(code = token)
                    data.loc[cmpt] = [intent , formulation]
                    cmpt = cmpt + 1
        return data
                    
                
        
#print(data["demande"])
tokens = ["INF3036" , "INF3046", "INF3176","INF3186","INF3196","INF3206","INF3216","INF3226","INF3236","INF3246","INF3266","INF3276"]                
new_tokens = ["INF142","INF112","INF122"]

g = generator()
g.fit(data , tokens)
test = g.generate(new_tokens)
```
